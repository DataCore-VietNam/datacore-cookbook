# DataCore Quickstart

Connect, search the catalog, and pull your first Vietnamese equity dataset in under 60 seconds.

**Prerequisites**: Set `DATACORE_API_KEY` in your environment. Get a free key at [datacore.vn/keys](https://datacore.vn/keys).

```python
from datacore import Client

dc = Client()  # reads DATACORE_API_KEY from env
domains = dc.list_domains()
domains
```

## Search the catalog

Hybrid keyword + semantic search ranks all 91 datasets by relevance.

```python
results = dc.search("VN30 daily prices", limit=5)
for r in results:
    print(r['id'], '—', r['name'])
```

## Pull a dataset as a DataFrame

```python
df = dc.dataset('equity.vn30.daily').to_pandas(
    start='2024-01-01',
    end='2024-12-31',
)
df.head()
```
